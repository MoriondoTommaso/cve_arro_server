```python
import pandas as pd 
from pathlib import Path
import numpy as np
```

```python
ROOT = Path.cwd().parent
DATA = ROOT / 'data' 
```

```python
df = pd.read_parquet(DATA / 'data' / 'cve_embs' /'cve_corpus.parquet')
df.head()
```

```python
cve99_to_14 = df[df['year'] <= 2014]

embs99_to_14 = cve99_to_14['embedding'].to_numpy()
np.save(DATA / 'cve_embeddings_demo' / 'embs_99_to_14.npy', embs99_to_14)
```

```python
cve99_to_14 = df[df['year'] > 2014]

embs99_to_14 = cve99_to_14['embedding'].to_numpy()
np.save(DATA / 'cve_embeddings_demo' / 'embs_99_to_14.npy', embs99_to_14)
```
